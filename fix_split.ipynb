```python
from casacore.tables import table, tablecopy, tablerename, tabledelete, taql
from pathlib import Path
import shutil
import numpy as np
import matplotlib.pyplot as plt
from fixms.fix_ms_corrs import get_pol_axis_coro
```

```python
in_path = Path("/scratch3/tho822/flint_oh/70731_bak")
ms_list = sorted(list(in_path.glob("*.ms")))
out_dir = Path("/scratch3/tho822/flint_oh/70731")
```

```python
for ms in ms_list:
    print(ms)
    new_ms = out_dir / ms.name
    if new_ms.exists():
        shutil.rmtree(new_ms)
    shutil.copytree(ms, new_ms)

    feed_path = new_ms / "FEED"


    with table(str(feed_path), readonly=False) as feed:
        RECEPTOR_ANGLE = feed.getcol("RECEPTOR_ANGLE")
        nrows = feed.nrows()
        row_idxs = np.arange(nrows)
        selected = (RECEPTOR_ANGLE == RECEPTOR_ANGLE[0]).sum(axis=1).astype(bool)
        feed.removerows(row_idxs[~selected])
        
```
